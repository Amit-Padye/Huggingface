```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
from transformers import pipeline
```

```python
rcParams['figure.figsize']=10,5
```

```python
zero_shot_classifier=pipeline('zero-shot-classification')
```

```python
result=zero_shot_classifier(sequences="A repository contains all project files, including the revision history. Already have a project repository elsewhere?t",candidate_labels=['data','data sciense','amit','food'],multi_class=True)
```

```python
result
```

```python
plt.bar(result['labels'],result['scores'])
plt.yticks(list(np.arange(0,1,0.1)))
plt.show()
```
